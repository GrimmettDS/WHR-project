```python
import os
import pandas as pd
```

```python
country_list = []
def list_country(l):
    for country in l:
        if country in country_list:
            continue
        else:
            country_list.append(country)
```

```python
list_country(list_2015)
list_country(list_2016)
list_country(list_2017)
list_country(list_2018)
list_country(list_2019)
list_country(list_2020)
```
